```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


def look3D(df, colPlot, texts, color_map,name):

    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    
    ax.scatter(
        df[colPlot[0]] ,
        df[colPlot[1]],
        df["score"],
        c=color_map, marker='o'
    )
    
    if(len(texts) != 0):
        for idx, row in pd.concat(texts).iterrows():
            x_coord =  df[colPlot[0]]
            y_coord =  df[colPlot[1]]

            ax.text(
                x_coord,
                y_coord,
                row['score'],
                row[name],
                fontsize=8
            )

    
    ax.set_xlabel(colPlot[0] + " = x")
    ax.set_ylabel(colPlot[1] + " = y")
    ax.set_zlabel("score")
    ax.set_title('주석이 포함된 3D 산점도')

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from sklearn.preprocessing import MinMaxScaler

# Set font to avoid issues with displaying Korean characters
plt.rcParams['font.family'] = 'Malgun Gothic'  # For Windows, use 'AppleGothic' for macOS
plt.rcParams['axes.unicode_minus'] = False

# Load the data from uploaded files
jinak_su_data_path = './data/education/유치원초중고 진학수및 취업수 통계 자료 2019_2024 (1).csv'
location_data_path = './data/education/초중고위치데이터_위도_경도.csv'
facility_data_path = './data/education/유치원초중고 시설정보 2019_2024.csv'
updated_facility_data_path = './data/education/유치원초중고 시설정보 2019_2024.csv'

jinak_su_data = pd.read_csv(jinak_su_data_path)
location_data = pd.read_csv(location_data_path)
facility_data = pd.read_csv(facility_data_path)
updated_facility_data = pd.read_csv(updated_facility_data_path)

# Step 1: Filter data for high schools and calculate ratio of students progressing to colleges and facility score
high_school_data = jinak_su_data[jinak_su_data['학제유형명'].str.contains('고등학교')]
high_school_data['진학 학생수'] = high_school_data[
    ['국내대학진학학생수', '국내전문대학진학학생수', '국내각종학교진학학생수', '국내교육대학진학학생수']
].sum(axis=1)

display(high_school_data.head(2))
high_school_data['진학률'] = high_school_data['국내대학진학학생수'] / high_school_data['진학 학생수']
high_school_facility_data = facility_data[facility_data['학교급명'].str.contains('고등학교')]
high_school_data = pd.merge(high_school_data, high_school_facility_data[['학교명', '교사_학교사옥면적', '체육공간수', '교사_도서관수']], how='left', on='학교명')
high_school_data['총시설점수'] = high_school_data[['교사_학교사옥면적', '체육공간수', '교사_도서관수']].sum(axis=1)


scaler = MinMaxScaler()
high_school_data['총시설점수'] = scaler.fit_transform(high_school_data[['총시설점수']])

orderhighCols = ['진학률', '총시설점수']
# Apply K-means clustering for high schools using both 진학률 and 총시설점수
kmeans_high = KMeans(n_clusters=3, random_state=42)
high_school_data[orderhighCols] = high_school_data[orderhighCols].fillna(0)
high_school_data['등급'] = kmeans_high.fit_predict(high_school_data[orderhighCols])


centroids = kmeans_high.cluster_centers_
# 각 점과 클러스터 중심 간의 거리 계산
high_school_data['DistanceToCenter'] = high_school_data.apply(
    lambda row: np.sum(abs(row[orderhighCols] - centroids[int(row['등급'])])),
    axis=1
)
#실제 해당 행의 거리 즉 최종 점수
high_school_data['score'] = high_school_data['DistanceToCenter'] + np.sum(centroids[high_school_data['등급']], axis=1)

# Step 2: Use facility information to grade middle schools with total facility score only

middle_school_data = facility_data[facility_data['학교급명'].str.contains('중학교')]
middle_school_data['총시설점수'] = middle_school_data[['교사_학교사옥면적', '체육공간수', '교사_도서관수']].sum(axis=1)  # Adding facility scores for clustering
kmeans_middle = KMeans(n_clusters=3, random_state=42)
middle_school_data['등급'] = kmeans_middle.fit_predict(middle_school_data[['총시설점수']].fillna(0))


# Step 3: Use facility information to grade elementary schools with total facility score only
elementary_data = facility_data[facility_data['학교급명'].str.contains('초등학교')]
elementary_data['총시설점수'] = elementary_data[['교사_학교사옥면적', '체육공간수', '교사_도서관수']].sum(axis=1)  # Adding facility scores for clustering
kmeans_elementary = KMeans(n_clusters=3, random_state=42)
elementary_data['등급'] = kmeans_elementary.fit_predict(elementary_data[['총시설점수']].fillna(0))


# Step 4: Merge location data with graded schools (high schools, middle schools, and elementary schools)
graded_schools = pd.concat([
    high_school_data[['학교명', '등급']],
    middle_school_data[['학교명', '등급']],
    elementary_data[['학교명', '등급']]
])

# Merge with location data to get latitude and longitude
final_data = pd.merge(graded_schools, location_data, how='left', on='학교명')
final_data = final_data[['학교명', 'lat', 'lon', '등급']]

# Exclude schools with missing latitude or longitude
final_data = final_data.dropna(subset=['lat', 'lon'])

# Save the final result to a CSV file
output_path = './export/education/final_graded_school_locations.csv'
final_data.to_csv(output_path, index=False)

print(f"CSV file saved to {output_path}")

# Plot distribution of grades
grade_counts = final_data['등급'].value_counts()
grade_counts.plot(kind='bar', color=['blue', 'green', 'red'])
plt.xlabel('등급')
plt.ylabel('학교 수')
plt.title('학교 등급 분포')
plt.show()



color_map = [colors[int(cls) - 1] for cls in high_school_data['등급']]  # 클래스가 1부터 시작하므로 -1

look3D(high_school_data, ['진학률', '총시설점수'], [], color_map,'학교명')

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)
colors = ['r','g','b']
# High school clustering distribution
axes[0].scatter(high_school_data['진학률'], high_school_data['총시설점수'], c=[colors[int(cls) - 1] for cls in high_school_data['등급']])
axes[0].set_xlabel('진학률')
axes[0].set_ylabel('총시설점수')
axes[0].set_title('고등학교 클러스터링 분포')


display(middle_school_data.describe())
# Middle school clustering distribution
axes[1].scatter(middle_school_data['총시설점수'], middle_school_data['등급'], c=[colors[int(cls) - 1] for cls in middle_school_data['등급']])
axes[1].set_xlabel('총시설점수')
axes[1].set_title('중학교 클러스터링 분포')

# Elementary school clustering distribution
axes[2].scatter(elementary_data['총시설점수'], elementary_data['등급'], c=[colors[int(cls) - 1] for cls in elementary_data['등급']])
axes[2].set_xlabel('총시설점수')
axes[2].set_title('초등학교 클러스터링 분포')

plt.tight_layout()
plt.show()
```
