```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/유치원초중고 진학수및 취업수 통계 자료 2019_2024.csv")
cols = df.columns
cutoff = df[df['조사년도'] > 2018]
cutoff = cutoff[cutoff['시도명'] == '서울']

data_columns = [
    "조사년도","개방ID","학교명","학교급명","시도명","학제유형명",

    "유초중등학생_학생수", "초등과정졸업생수","중등과정졸업생수","고등과정졸업생수", "유초중등학생_졸업생수",
    
    "취업자수",
    
    "일반고진학학생수", "특수목적고등학교진학학생수","자율고등학교진학학생수","특성화고진학학생수", "특수학급학생수","진학학생수",
    
    "국내대학진학학생수", "국내전문대학진학학생수", "국내각종학교진학학생수","국내교육대학진학학생수"
]
cutoff = cutoff[data_columns]


display(cutoff[cutoff['특수목적고등학교진학학생수'] != 0])

cutoff.to_csv('../export/education/output2.csv', index=False, encoding='utf-8')  # index=False로 행 번호 제외

for now in cols:
    print(now)
```
