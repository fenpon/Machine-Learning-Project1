```python
#경찰 데이터 변환

import pandas as pd
import numpy as np
```

```python
polices = pd.read_csv('../data/safety/policeFacilities.csv')
polices = polices[polices['시도청'] == '서울청']

mapper = {'파출소':2,'지구대':1}

polices = polices[['경찰서','관서명','구분','주소']]
polices['구분'] = polices['구분'].map(mapper)
polices['이름'] =  polices['경찰서'] +" "+  polices['관서명']
polices = polices.drop(columns=['경찰서','관서명'])
polices.head()
```

```python
polices.to_csv('../export/safety/policeFacilities.csv', index=False, encoding='utf-8')  # index=False로 행 번호 제외
```
